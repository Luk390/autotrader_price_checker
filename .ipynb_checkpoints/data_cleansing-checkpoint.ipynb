```python
import pandas as pd
import missingno as msno
import numpy as np
from matplotlib import pyplot as plt
```

```python
raw_data = pd.read_csv('df.csv', header=0)
```

```python
raw_data.head()
```

```python
df = raw_data.drop(['description', 'emailAddress', 'raw_response', 'status_code', 'tax'], axis=1)
df.head()
```

```python
#check categorical variables - body_type, condition, fuel_type, isTradeSeller, make, model, townAndDistance, transmission, trim
for column in ['body_type', 'condition', 'fuel_type', 'isTradeSeller', 'make', 'model', 'townAndDistance', 'transmission', 'trim']:
    print(df[column].value_counts().sort_index())
    print('\n')
```

```python
#Drop Pickup and Limosine due to low values
df.drop(df[(df['body_type'] == 'Limousine') | (df['body_type'] == 'Pickup')].index, inplace=True)


#Drop CUPRA, Lexus and Subarus due to low sample numbers
df.drop(df[(df['make'] == 'Subaru') | (df['make'] == 'CUPRA') | (df['make'] == 'Lexus')].index, inplace=True)

#After confirming all the cars are indeed petrol, this column can be dropped
df.drop(['fuel_type'], axis=1, inplace=True)

#The data was scraped as 10 miles from the center of Bristol, London, Birmingham, Leeds and Manchester. 
#So will assign these values to a new column based on the postcode used in url

location_dict = {'bs11aa': 'Bristol', 
                 'wc2n5du': 'London', 
                 'b11bb': 'Birmingham', 
                 'ls11az': 'Leeds', 
                 'm11ag': 'Manchester'}
def location_url(string):
    for key in location_dict.keys():
        if key in string:
            return location_dict[key]

df['location'] = df.url.map(location_url)

#Drop url and townAndDistance columns as no longer needed
df.drop(['url', 'townAndDistance'], axis=1, inplace=True)

#Trim L of engine_size
df['engine_size'] = df['engine_size'].apply(lambda x: str(x)[:-1])

#Replace string 'na' with np.Nan
df['engine_size'] = df['engine_size'].apply(lambda x: np.NaN if x == 'na' else x)

#Drop column trim due to over 60% values missing.
#This is not so much of a problem because there were over 300 unique trim options, so would have drastically upped the 
#number dummy variables needed
df.drop(['trim'], axis=1, inplace=True)

#Fix a specific instance where one record's engine size was 1000 times what is should have been
df.engine_size = df.apply(lambda x: 0.9 if x.engine_size == '875.0' else x.engine_size, axis=1)


```

```python
msno.bar(df)
```

```python
#I decided that I due to the relatively low number of records which have missing values. 
#I would drop a row if there was any missing values. This drops the total records from 5842 to 5525
df = df.dropna()
```

```python
msno.bar(df)
```

```python
df.dtypes
```

```python
for column in ['body_type', 'condition', 'isTradeSeller', 'make', 'model',
               'transmission', 'location', 'doors', 'manufactured_year', 'seats']:
    df[column] = df[column].astype('category')
    
for column in ['co2Emissions', 'mileage']:
    df[column] = df[column].astype('int64', copy=False)

df.engine_size = df.engine_size.astype('float64')
```

```python
df.dtypes
```

```python
for column in df:
    if (df[column].dtype == 'int64') | (df[column].dtype == 'float64'):
        plt.boxplot(df[column])
        plt.title(f"Boxplot for {column}")
        plt.show()
```

The boxplots for the numerical datatypes show that most are skewed quite far to the right. The assumptions for the linear regression assume that the residuals are normally distributed, so a lack of Gaussian distribution shouldn't pose a problem. However there is an extreme outlier in price. This needs investigating to confirm the validity of the data point.

```python
df[df['price'] > 800000]
```

This certainly explains the outlier. It is a one off commission by McLaren and it is beautiful. Despite this being a genuine data point, it will be removed from the dataset. As this data point is for a one off commission, it holds very little value in being able to predict the price of almost all cars. So leaving it in would skew the results of the prediction with little value, as much of the value of this car is down to its uniqueness, which is not captured in the variables in the dataframe.

```python
#Remove the McLaren Senna from the dataset
df.drop(df[df['price'] > 800000].index, inplace=True)
```

```python
df.to_csv('data.csv')
```
